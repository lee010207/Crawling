### 하나의 셀에 코드 합치기
- 프로그램(실행파일) 제작을 할 때는 반드시 코드를 하나의 셀에 합쳐야 함

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
time.sleep(3)

try : 
    for i in range(15):
        btn = driver.find_element(By.CSS_SELECTOR,".c_05")
        btn.click()
        # 요청을 하고 응답이 오는 동안 약간의 시간을 주겠다 (요청과 응담 사이에)
        time.sleep(3)
except :
    print("더보기가 없습니다!")
# -- 데이터 준비 단계 --

menu = driver.find_elements(By.CSS_SELECTOR,".h.fz_03")
price = driver.find_elements(By.CSS_SELECTOR,".item-price>strong")
# -- 데이터 수집 단계 --

menu_list = []
price_list = []
for i in range(len(menu)):
    menu_list.append(menu[i].text)
    price_list.append(price[i].text)
# -- 데이터 가공 단계 --

dic = {"상품명" : menu_list, "가격정보" : price_list}
hansot = pd.DataFrame(dic)
hansot.to_csv('한솥.csv', encoding='euc-kr')
# -- 데이터 활용 단계 --
```

# 크롤링 프로세스
    1. 데이터 준비 : 완벽한 데이터를 수집하기 위해 화면 준비를 하는 단계
       - 더보기 버튼을 모두 클릭해서 페이지를 전체를 한 화면에 담아놓기, 화면 스크롤을 최하단까지 이동시키기 
    2. 데이터 수집 : 준비된 데이터(html)에서 필요한 요소를 수집
    3. 데이터 가공 : 수집된 데이터(요소)를 필요한 컨텐츠, 속성만 뽑아내는, 가공하는 과정
    4. 데이터 활용 : 가공된 데이터를 표, 파일로 제작
  

# 크롤링 팁
- 반드시 selenium을 활용할 때는 '셀 런 올' 금지!
- 항상 사람처럼 코딩! : 요청과 응답 사이에는 반드시 시간을 부여
- driver.implicitly_wait(초) : 요청 - 응답 통신이 설정한 시간 이내에 끝나면 안기다리고 바로 코드 실행됨

        tip!
        time.sleep(): 무조건 내가 설정한 시간만큼 코드를 중단 -> 무조건 기다림
        implicitly_wait(): 설정한 시간보다 완료된 시간이 짧으면 완료가 된 기점으로 코드가 진행
            - 요청과 응답 통신이 빠를 때, 네트워크 환경이 좋을 때는 장점(빠름)
            - 네트워크 환경이 좋지 않을 때는 데이터가 완벽하게 안넘어올 수 도 있음(권장X)

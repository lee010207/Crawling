### 리뷰데이터 가져오기
    한 페이지 안에 보여줄 데이터가 많아지면 새로고침할 때마다 자원 소모가 크므로 각 부분별로 페이지를 따로 만들어서 한페이지에 보여주기만 하는 형태로 구성된 경우
    
    - iframe : 기존 페이지 안에 새로운 페이지를 삽입하는 태그
    - 기존 페이지 안에서는 iframe 안에 있는 값(다른 페이지)을 가져올 수 없음
    - iframe 안에 있는 src속성의 url에 접근해서 연결된 해당 페이지(리뷰 페이지)를 크롤링해와야 내가 원하는 데이터를 가져올 수 있음!
    
    <방법>
    (1) 개발자 도구 -> ctrl + f : iframe 검색
    (2) 원하는 부분의 iframe 태그 찾기
    (3) 해당 iframe태그 안의 src 속성 url 가져와서
    (4) req.get()안에 넣어주기

```python
# 리뷰를 수집할 때 많이 발생 : 여러개의 페이지로 구성되어 있을 때
# > 한 페이지 안에서 원하는 값이 여러개의 페이지로 구성되어 있는 경우
# > 새로고침 할때마다 많은 자원이 소모됨 => 하나의 페이지로 만들어서 보여주기만
# review = soup.select("div.review_list_v2__message.js-collapsed-review-content.js-translate-text")

review = soup.select("div.review_list_v2__expand_link.js-renewal-review-message-link.js-renewal-link-expand>div.review_list_v2__message.js-translate-text")
```

```python
# 별점
rating = soup.select("div.review_list_v2__score_star > span")
```

```python
#작성자 user
user = soup.select("div.review_list_v2__user_name_message>b")
user
```

```python
import pandas as pd
```

```python
# 이스케이프 코드 : \n,
# strip() : 문자열 앞,뒤의 공백 제거
```

```python
User = []
Rating = []
Review = []

for i in range(len(user)):
    User.append(user[i].text)
    Rating.append(rating[i].text)
    Review.append(review[i].text)
#    Review[i] = Review[i].strip()

cnt = 1    
for i in Review:
    print(cnt)
    print(i.strip())
    cnt+=1

```

```python
dic = {'작성자': User,'별점':Rating,'리뷰':Review}
data = pd.DataFrame(dic)
data
```
